```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
path = 'd:/Dataset/'
imgpath1 = path + '4.2.01.tiff'
imgpath2 = path + '4.2.03.tiff'
```

```python
img1 = cv2.imread(imgpath1, 1)
img2 = cv2.imread(imgpath2, 1)
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
mult = img1 * img2
div = img1 / img2
```

```python
titles = ['Liquid Drop', 'Mandrill', 'Multiplication', 'Division']
```

```python
images = [img1, img2, mult, div]

for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
alpha = 0.5
beta = 0.5
gamma = 0

# img1 * alpha + img2 * beta + gamma

output = cv2.addWeighted(img1, alpha, img2, beta, gamma)

titles = ['Liquid Drop', 'Mandrill', 'Weighted Addition']
images = [img1, img2, output]
```

```python
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
import time

path = 'd:/Dataset/'

path = 'd:/Dataset/'
imgpath1 = path + '4.2.01.tiff'
imgpath2 = path + '4.2.03.tiff'

img1 = cv2.imread(imgpath1, 1)
img2 = cv2.imread(imgpath2, 1)

for i in np.linspace(0, 1, 1000):
    alpha = i
    beta = 1 - alpha
    output = cv2.addWeighted(img1, alpha, img2, beta, 0)
    cv2.imshow('Transition', output)
    time.sleep(0.0001)
    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()
```

```python
def emptyFunction():
    pass
```

```python
output = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
windowName = 'Transition Demo'
```

```python
cv2.namedWindow(windowName)
cv2.createTrackbar('Alpha', windowName, 0, 1000, emptyFunction)

while True:
    cv2.imshow(windowName, output)
    if cv2.waitKey(1) == 27:
        break
    alpha = cv2.getTrackbarPos('Alpha', windowName) / 1000
    beta = 1 - alpha
    output = cv2.addWeighted(img1, alpha, img2, beta, 0)
    print(alpha, beta)
cv2.destroyAllWindows()
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
r, g, b = cv2.split(img1)

titles = ['Original', 'Red', 'Green', 'Blue']

images = [cv2.merge((r, g, b)), r, g, b]
```
